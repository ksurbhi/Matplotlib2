# Problem 1: Line chart in Matplotlib


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.array([1,2,3,4,5])
y = X*2

# Create a line chart
plt.plot(X,y,'-.',marker = 'o',c ='red')

# Add annotations
for i,(xi,yi) in enumerate(zip(X,y)):
  plt.annotate(f'({xi},{yi})',(xi,yi), textcoords='offset points',xytext=(0, 10), ha='center')
  # plt.annotate(f'({xi}, {yi})', (xi, yi), textcoords="offset points", xytext=(0, 10), ha='center')

# Add title and labels
plt.title('My First Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
# Display grid
plt.grid(True)

# Show the plot
plt.show()

# The figure() function helps in creating a
# new figure that can hold a new chart in it.
plt.figure()

# Second graph
x1 = np.array([3,5,7,9,11])
y1 = x1 * 2

plt.plot(x1,y1,'-',marker = 'o',c ='blue')
# Add title and labels for 2nd graph
plt.title('My 2nd graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# add annotations
for i, (xi, yi) in enumerate(zip(x1,y1)):
  plt.annotate(f'({xi},{yi})',(xi,yi), textcoords= 'offset points', xytext= (0,10),ha = 'center')
# Display grid
plt.grid(True)
# Display the plot
plt.show()
```

# Problem 2: Plotting Histogram in Python using Matplotlib


```python
# Basic Histogram
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(234567)
X = np.random.randn(1000)
plt.hist(X, bins = 30, color = 'pink',edgecolor = 'black')
plt.xlabel('X-values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')
plt.show()
```

```python
# histogram with density
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
np.random.seed(234567)
X = np.random.randn(1000)
sns.histplot(X, bins = 30, color = 'skyblue', edgecolor = 'black',kde = True)
plt.xlabel('X-values')
plt.ylabel('Frequency')
plt.title('Histogram with density')
plt.show()
```

```python
# Customized Histogram with Watermark
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors
np.random.seed(2346789)
X = np.random.randn(10000)
y = 8 ** X + np.random.randn(10000) + 25

# Creating histogram
fig, axs = plt.subplots(1,1, figsize = (10,7),tight_layout = True)
# Remove axes splines
for s in ['top','bottom','left','right']:
  axs.spines[s].set_visible(False)


# Remove x, y ticks
axs.xaxis.set_ticks_position('none')
axs.yaxis.set_ticks_position('none')

# Add padding between axes and labels
axs.xaxis.set_tick_params(pad= 5)
axs.yaxis.set_tick_params(pad =10)

# Add x, y gridlines
axs.grid(color = 'grey',linestyle = '-.')

# Add Text watermark
fig.text(0.9, 0.15, 'JMy customize histogram',
         fontsize = 12,
         color ='red',
         ha ='right',
         va ='bottom',
         alpha = 0.7)

# Creating histogram
N, bins, patches = axs.hist(X, bins = 20)

# Setting color
fracs = ((N**(1 / 5)) / N.max())
norm = colors.Normalize(fracs.min(), fracs.max())

for thisfrac, thispatch in zip(fracs, patches):
    color = plt.cm.viridis(norm(thisfrac))
    thispatch.set_facecolor(color)
# Adding extra features
legend = ['distribution']
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.legend(legend)
plt.title('Customized histogram')

plt.show()
```

```python
#Multiple Histograms with Subplots
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for multiple histograms
data1 = np.random.randn(1000)
data2 = np.random.normal(loc=3, scale=1, size=1000)

# Creating subplots with multiple histograms
fig,axs = plt.subplots(1,2,figsize =(10,10))

# Plot histoframs
axs[0].hist(data1, color = 'skyblue',edgecolor = 'black',bins =30)
axs[1].hist(data2, color = 'pink',edgecolor = 'black',bins =30)

# Title and labels
axs[0].set_title('Histogram-1')
axs[1].set_title('Histogram-2')

axs[0].set_xlabel('values')
axs[0].set_ylabel('Frequency')
axs[1].set_xlabel('values')
axs[1].set_ylabel('Frequency')

# Adjusting layout for better spacing
plt.tight_layout()
# Display graph
plt.show()
```

# Problem 3: Matplotlib Scatterplot in Matplotlib

```python
# Basic scatter plot
import matplotlib.pyplot as plt

x =[5, 7, 8, 7, 2, 17, 2, 9,
    4, 11, 12, 9, 6]

y =[99, 86, 87, 88, 100, 86,
    103, 87, 94, 78, 77, 85, 86]

plt.scatter(x,y,color = 'red')
# Title and Labels
plt.title("Basic Scatter Plot")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Display the plot
plt.show()
```

```python
# Bubble Plots in Matplotlib


import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y= [2, 3, 5, 7, 11]

# Bubble sizes
b_size = [50,100,150,200,300]
# Create bubble chart
plt.scatter(x, y, s = b_size, c = 'skyblue',alpha=0.7, edgecolors='blue', linewidths=2)

# Title and Labels
plt.title('Bubble plot on a Scatterplot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Display the plot
plt.show()
```

```python
# Customized pie chart
import matplotlib.pyplot as plt
import numpy as np

# Dataset
cars = ['mazda','BMW','Mercedes','Audi','Tesla','Ford']

data = [23, 17, 35, 29, 12, 41]
# Explode data
explode = (0.1,0.0,0.2,0.3,0.0,0.0)
# wedge properties
wp = {'linewidth': 1, 'edgecolor' : 'green'}
# using customized colors
colors = ['orange','cyan','brown','grey','indigo','beige']
plt.pie(x = data,labels = cars, explode =explode, autopct='%1.1f%%',
        shadow = True, startangle= 90,colors= colors,wedgeprops=wp,
        textprops= dict(color ='magenta'))
#adding legends
plt.legend(title = 'cars', loc = 'center left',bbox_to_anchor=(1, 0, 0.5, 1))

plt.title("Customized pie chart")
plt.show()
```

```python
# Nested pie chart
# Import libraries
from matplotlib import pyplot as plt
import numpy as np


# dataset
size = 6
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES']

data = np.array([[23, 16], [17, 23],
                 [35, 11], [29, 33],
                 [12, 27], [41, 42]])

# normalizing data to 2 pi
norm = data / np.sum(data)*2 * np.pi

# obtaining ordinates of bar edges
left = np.cumsum(np.append(0,
                           norm.flatten()[:-1])).reshape(data.shape)

# Creating color scale
cmap = plt.get_cmap("tab20c")
outer_colors = cmap(np.arange(6)*4)
inner_colors = cmap(np.array([1, 2, 5, 6, 9,
                              10, 12, 13, 15,
                              17, 18, 20]))

# Creating plot
fig, ax = plt.subplots(figsize=(10, 7),
                       subplot_kw=dict(polar=True))

ax.bar(x=left[:, 0],
       width=norm.sum(axis=1),
       bottom=1-size,
       height=size,
       color=outer_colors,
       edgecolor='w',
       linewidth=1,
       align="edge")

ax.bar(x=left.flatten(),
       width=norm.flatten(),
       bottom=1-2 * size,
       height=size,
       color=inner_colors,
       edgecolor='w',
       linewidth=1,
       align="edge")

ax.set(title="Nested pie chart")
ax.set_axis_off()

# show plot
plt.show()
```

# Pie chart for Churn Modelling Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('Churn_Modelling.csv')
df = dataset[dataset['Gender'] == 'Female']
df = df[df['HasCrCard']==1]
df = df.groupby('Geography').size().reset_index(name = 'count')
# colors
colors = ['pink','skyblue','beige']
#wedgw properties
wp = {'linewidth':1,'edgecolor':'green'}
# explode
explode = (0.0,0.1,0.0)
# creating pie chart
plt.pie(x = df['count'],labels = df['Geography'],explode = explode,
        colors = colors,shadow = True, autopct = '%1.1f%%',
        startangle = 90, counterclock = False, wedgeprops= wp,
        textprops = dict(color = 'Magenta'))
# legends
plt.legend(title = 'Region', loc = 'center left',bbox_to_anchor = (1.1,0,0.5,1))
#title
plt.title('Pie plot for Females for each region with one credit card')
plt.show()
```
